### Statistical plotting with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

### Regression Plots

#### Linear Regression Plot

```python
sns.set_style("whitegrid")
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"s": 20}, line_kws={"color": "red"})
plt.title('Linear Regression Plot')
```

#### Multiple Linear Regression Plot

```python
sns.set_style("whitegrid")
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips, palette="YlGnBu")
plt.title('Multiple Linear Regression Plot')
```

#### Residual Plot

```python
sns.residplot(x="total_bill", y="tip", data=tips, scatter_kws={"s": 20})
plt.title('Residual Plot')
```

### Distribution Plots

```python
sns.set_style("whitegrid")
sns.histplot(tips["total_bill"], kde=False, color="blue")
plt.title('Histogram')
```

```python
sns.set_style("whitegrid")
sns.kdeplot(tips["total_bill"], color="blue", shade=True)
plt.title('Kernel Density Estimate Plot')
```

```python
sns.set_style("whitegrid")
sns.histplot(tips["total_bill"], kde=True, color="blue")
plt.title('Distribution Plot')
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
plt.title('Box Plot')
```

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu", split=True)
plt.title('Violin Plot')
```

```python
sns.pairplot(tips, hue="sex", palette="YlGnBu")
plt.title('Pair Plot')
```

```python
sns.set_style("whitegrid")
sns.kdeplot(tips["total_bill"], color="blue", fill=True)
plt.title('Kernel Density Estimate Plot')
```

```python
sns.set_style("whitegrid")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter", color="blue")
plt.title('Joint Plot')
```

```python
g = sns.FacetGrid(tips, col="sex", row="time", margin_titles=True, palette="YlGnBu")
g.map(sns.histplot, "total_bill")
```
