### Introduction to Matplotlib (basic plots: line, scatter, bar)

```python
import matplotlib.pyplot as plt
```

##                                                   Basic Plots

#### Create a Simple Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

#### Plot multiple lines

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()
plt.show()
```

#### Line plot with markers

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Markers')
plt.show()
```

#### Custom line style and color

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, linestyle='--', color='r', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Custom Line Style and Color')
plt.show()
```

## Scatter Plots

#### Create a scatter plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Scatter Plot')
plt.show()
```

#### Custom scatter plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, marker='x', s=100, c='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Custom Scatter Plot')
plt.show()
```

#### Multiple scatter plots

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]

plt.scatter(x1, y1, marker='o', c='blue', label='Dataset 1')
plt.scatter(x2, y2, marker='x', c='red', label='Dataset 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Scatter Plots')
plt.legend()
plt.show()
```

#### Scatter plot with labels and titles

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Labels and Titles')
plt.show()
```

## Bar Charts

#### Vertical bar chart

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [4, 7, 1, 8, 5]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart')
plt.show()
```

#### Horizontal bar chart

```python
plt.barh(categories, values)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.show()
```

#### Custom bar colors and edge colors

```python
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Custom Bar Colors')
plt.show()
```

## Pie Charts

#### Basic pie chart

```python
sizes = [20, 30, 25, 25]
labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4']

plt.pie(sizes, labels=labels)
plt.title('Basic Pie Chart')
plt.show()
```

#### Custom slice colors and legend

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors)
plt.legend(labels, loc="best")
plt.title('Custom Pie Chart with Legend')
plt.show()
```

#### Explode a slice

```python
explode = (0, 0.1, 0, 0)  # explode 2nd slice
plt.pie(sizes, labels=labels, explode=explode, colors=colors)
plt.title('Pie Chart with Exploded Slice')
plt.show()
```

#### Add percentage labels

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Pie Chart with Percentage Labels')
plt.show()
```
