### Advanced plotting techniques

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Subplots:

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]
data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Scatter plot
axs[0, 1].scatter(x, y2, color='r')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

# Bar plot
axs[1, 0].bar(categories, values, color='skyblue', edgecolor='black')
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')

# Histogram
axs[1, 1].hist(data, bins=30, color='g', edgecolor='black')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Histograms:

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  # shifted data

# Multiple histograms
plt.hist(data1, bins=20, alpha=0.5, label='Data 1')
plt.hist(data2, bins=20, alpha=0.5, label='Data 2')
plt.title('Multiple Histograms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Normalized histograms
plt.hist(data1, bins=20, alpha=0.5, density=True, label='Data 1')
plt.hist(data2, bins=20, alpha=0.5, density=True, label='Data 2')
plt.title('Normalized Histograms')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Box Plots

```python
data = np.random.randn(100)

plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

```python
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 2
data3 = np.random.randn(100) - 2

# Multiple box plots
data = [data1, data2, data3]
plt.boxplot(data)
plt.title('Multiple Box Plots')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.xticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])
plt.show()
```

```python
# Customized box plots
plt.boxplot(data, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            medianprops=dict(color='red'))
plt.title('Customized Box Plots')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.xticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])
plt.show()
```
