```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
sumis = np.add(array1, array2)
print(sumis)
```

```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
product = np.multiply(array1, array2)
print(product)
```

```python
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
dot_product = np.dot(matrix1, matrix2)
print(dot_product)
```

```python
arr = np.array([1, 2, 3, 4, 5])
mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```

```python
# cumulative sum
arr = np.array([1, 2, 3, 4, 5])
sumis = np.cumsum(arr)
print(sumis)
```

```python
arr= np.random.randint(1, 100, size=(5, 5))
correlation_matrix = np.corrcoef(arr, rowvar=False)
print(correlation_matrix)
```

```python
# Simulate rolling the die 1000 times
rolls = np.random.randint(1, 7, size=1000)
counts = np.bincount(rolls) # count frequency

for outcome, frequency in enumerate(counts[1:], start=1):
    print(f"Outcome {outcome}: {frequency} times")
```
