### Seaborn Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

### Scatter Plot

#### used palette here

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips, size = "size", hue = "day",palette = "YlGnBu") 
plt.title("Tips size paid per day")
```

### Histogram

#### Kde for line in graph

```python
sns.histplot(tips["tip"], kde = True, bins = 15) 
plt.title("Tip counts")
```

### Bar Plot

```python
sns.barplot(x = "sex", y = "tip", data = tips, palette = "YlGnBu")
sns.set_style("whitegrid")
plt.title("Tips by gender")
```

### Box plot

Middle line inside box is the median. Max and min are lines above and below box respectively and the dots above are outliers

```python
sns.boxplot(x = "day", y = "tip", data = tips, hue = "sex", palette = "YlGnBu")
sns.set_style("whitegrid")
plt.title('Tips given in days by males and females')
```

### Line Plot

```python
sns.lineplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu", marker="o")
sns.set_style("whitegrid")
plt.title('Average Tip by Day and Sex')
```

```python
sns.countplot(x="day", data=tips, hue="sex", palette="YlGnBu")
sns.set_style("whitegrid")
plt.title('Count of Tips by Day and Sex')
```
