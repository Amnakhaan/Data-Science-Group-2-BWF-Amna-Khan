### File Handling, Exception Handling


##### Add Students in file

```python
def add_students(student_id, name, grade):
    file = open("students.txt", 'a')
    file.write(f"{student_id}, {name}, {grade}\n")
    file.close()
    print("Student added successfully.")
```

#### View Student's Data

```python
def view_students():
    if open("students.txt"):
        file = open("students.txt", 'r')
        students = file.readlines()
        
        for student in students:
            student_data = student.strip().split(',')
            print(f"ID: {student_data[0]}\nName: {student_data[1]}\nGrade: {student_data[2]}\n")
    else:
        print("No file Exists")
```

#### Update Student's Data

```python
import os

def update_student(student_id, name=None, grade=None):
    if os.path.exists("students.txt"): #check if exists
        updated = False

        # Read file
        with open("students.txt", 'r') as file:
            students = file.readlines()
        
        # Write updates to file
        with open("students.txt", 'w') as file:
            for student in students:
                student_data = student.strip().split(',')
                if student_data[0] == student_id:  # Entered ID to update data
                    if name:
                        student_data[1] = name
                    if grade:
                        student_data[2] = grade
                    updated = True
                file.write(','.join(student_data) + '\n')
        
        if updated:
            print("Student data updated successfully.")
        else:
            print("Student ID not found.")
    else:
        print("No students found.")
```

#### Delete Student's Data

```python
def delete_student(student_id):
    if os.path.exists("students.txt"): #check if exists
        delete = False

        # Read file
        with open("students.txt", 'r') as file:
            students = file.readlines()
        
        with open("students.txt", 'w') as file:
            for student in students:
                student_data = student.strip().split(',')
                if student_data[0] != student_id: 
                    file.write(','.join(student_data) + '\n')
                else:
                    delete = True
        
        if delete:
            print("Student data Deleted successfully.")
        else:
            print("Student ID not found.")
    else:
        print("No students found.")
```

##### Main Function to run Student Management System

```python
def main():
    while True:
        
        print("\n\n------ Student Management System -------")
        print("1. Add Student")
        print("2. View Students")
        print("3. Update Student")
        print("4. Delete Student")
        print("5. Exit")

        choice = input("\nEnter your choice: ")

        if choice == '1':
                student_id = input("Enter student ID: ")
                name = input("Enter student name: ")
                grade = input("Enter student grade: ")

                add_students(student_id, name, grade)


        elif choice == '2':
            view_students()

        elif choice == '3':
            student_id = input("Enter student ID to update: ")
            name = input("Enter new name (leave blank to keep unchanged): ")
            grade = input("Enter new grade (leave blank to keep unchanged): ")

            update_student(student_id, name if name else None, grade if grade else None)

        elif choice == '4':
            student_id = input("Enter student ID to delete: ")

            delete_student(student_id)

        elif choice == '5':
            print("Exiting the system.")
            break # out of the while loop


        else:
            print("Invalid choice. Please try again.")
```

```python
if __name__ == "__main__":
    main()
```
